```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/airbnb_nyc_clean.csv')
```

```python
df.shape
```

```python
df.isna().sum()/df.shape[0]*100
```

```python
df.dropna(subset=['service_fee'])
```

```python
df.describe()
```

```python
#genero una columna para poder generar la definiciÃ³n binaria de ser mayor o igual a 4 o no.
df['supera_4']=np.where(df['review_rate_number'] >= 4, '>=4', '<4')
```

```python
#Armo grÃ¡fica de barras para ver si hay variaciÃ³n del rating con respecto al aÃ±o del alojamiento.
df_agrupada_cy_s4 = df.groupby(['construction_year', 'supera_4'])['supera_4'].count().unstack()
df_agrupada_cy_s4.plot(kind='bar', figsize=(10, 6))

plt.xlabel("AÃ±o de ConstrucciÃ³n")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de rating >=4 por AÃ±o de ConstrucciÃ³n")
plt.legend(title="Supera 4")
plt.xticks(rotation=45)

plt.show()
```

```python
#Armo grÃ¡fica de barras para ver si hay variaciÃ³n del rating con respecto al tipo de habitaciÃ³n.
df_agrupada_rt_s4 = df.groupby(['room_type', 'supera_4'])['supera_4'].count().unstack()
df_agrupada_rt_s4.plot(kind='bar', figsize=(10, 6))

plt.xlabel("Tipo de habitaciÃ³n")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de rating >=4 por Tipo de habitaciÃ³n")
plt.legend(title="Supera 4")
plt.xticks(rotation=45)

plt.show()
```

```python
#Armo grÃ¡fica de barras para ver si hay variaciÃ³n del rating con respecto a la ciudad.
df_agrupada_ng_s4 = df.groupby(['neighbourhood_group', 'supera_4'])['supera_4'].count().unstack()
df_agrupada_ng_s4.plot(kind='bar', figsize=(10, 6))

plt.xlabel("Ciudad")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de rating >=4 por ciudad")
plt.legend(title="Supera 4")
plt.xticks(rotation=45)

plt.show()
```
